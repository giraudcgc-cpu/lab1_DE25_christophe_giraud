```python
# WELCOME TO LAB1 - Task 1 

"""
This is my 2nd and final version for task1. 

The first attempt is called seq1.ipynb and I left it for you to see where I started from. 
The code was very long with plenty of redundancies like many "if-statements" as I went 
through each sequence with the same "process". Then, I tried to use functions thinking to "group things that
can be used together" but I realized at the stuga that one would have to call the functions which is not 
user friendly.

Ideas found online/LLM:
.lstrip()
http://realpython.com/videos/strip-characters-from-string-overview/
https://www.w3schools.com/python/ref_string_lstrip.asp

To check my code or get hints, I used LLM but it shows often fixes that I never heard of or felt not confident to use.
For example, we saw the method ".strip()" but I did not know ".lstrip()". 

Bar charts: 
https://matplotlib.org/stable/gallery/lines_bars_and_markers/bar_colors.html#sphx-glr-gallery-lines-bars-and-markers-bar-colors-py

Helped from Stuga time (more for task 2)
"""


import matplotlib.pyplot as plt

with open("dna_raw.txt", "r") as f:         # Opening and reading the file, with-statement
    data = f.read()
    print(f"The printed text (the original):\n {data}")

lines = data.lower().split("\n")            # converting all to lowercase & splitting into lines with no white spaces


seq1_n = lines[0].lstrip(">")               # split lines, naming, index and stripping the symbol on the left
seq2_n = lines[2].lstrip(">")
seq3_n = lines[4].lstrip(">")
seq4_n = lines[6].lstrip(">")

seq1_letters = lines[1]                     # split lines per serie of letters/sequence
seq2_letters = lines[3]
seq3_letters = lines[5]
seq4_letters = lines[7]


all_sequences = {                            # created dictionary to store the names/ids and the letters. Felt more confident using dict.
    seq1_n: seq1_letters,
    seq2_n: seq2_letters,
    seq3_n: seq3_letters,
    seq4_n: seq4_letters
}
print()
print()
print("-------DNA Sequence Data: Dictionaries and Bar Charts-------")
print()

for seq_n, letters in all_sequences.items():  # Looping in the dict via the keys and values 
    letter_counts = {"a": 0, "c": 0, "g": 0, "t": 0}   # Setting a dict to count valid letters, excluding "N". At the beginning I used many if-statements, was too heavy to read
    for letter in letters:
        if letter in letter_counts:
            letter_counts[letter] += 1                  # incrementing values 
    
    print(f"{seq_n} = {letter_counts}")                 # Printing results

    
    fig, ax = plt.subplots()                            # bar chart creation, removed a function that I used at first after talking at the stuga
    dna_letters = list(letter_counts.keys())
    frequency = list(letter_counts.values())
    bar_colors = ['tab:red']
    ax.bar(dna_letters, frequency, color=bar_colors)
    plt.xlabel("DNA Letters")
    ax.set_ylabel("Frequency")
    ax.set_title(f"DNA {seq_n}")
    plt.show()

```
